### Read an example image.


```python
img = cv2.imread('daisy.png')
cv2_imshow(img)
```

### Line Segmentation

```python
sobelx = cv2.Sobel(src=img, ddepth=cv2.CV_64F, dx=1, dy=0, ksize=3) # Sobel Edge Detection on the X axis
sobely = cv2.Sobel(src=img, ddepth=cv2.CV_64F, dx=0, dy=1, ksize=3) # Sobel Edge Detection on the Y axis
```

### Display Result

```python
cv2_imshow(sobelx)
```

This image represent horizontal Sobal derivative. It is show the border on left most of object.

```python
cv2_imshow(sobely)
```

This image represent vertical Sobal derivative. It is show the border on top of object.

```python
cv2_imshow(np.sqrt(sobelx**2+sobely**2))
```

I combine the two previous results. So we can approximation of the gradient that represent border of object in image.We can see white border of daisy flower and wood line.And we can see blue and yellow on middle of flower, Because we calculate sobel on rgb image.
